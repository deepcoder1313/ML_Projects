```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_samples,silhouette_score
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
```

```python
df =pd.read_csv("netflix.csv",encoding = "latin1")
df.head()
```

```python
df= df.drop_duplicates()
df.duplicated().sum()
```

```python
df = df.drop(["id","age_certification","imdb_id","tmdb_popularity","tmdb_score"],axis =1)
df.tail()
```

```python
df["production_countries"] = (
    df["production_countries"]
    .str.replace(r"\[", "", regex=True)
    .str.replace(r"'", "", regex=True)
    .str.replace(r"\]", "", regex=True)
)
df["lead_production"]=df["production_countries"].str.split(",").str[0]
df['production_countries_cnt'] = df['production_countries'].str.split(',').str.len()
df['production_countries_cnt']
```

```python
df["genres"] = (
    df["genres"]
    .str.replace(r"\[", "", regex=True)
    .str.replace(r"'", "", regex=True)
    .str.replace(r"\]", "", regex=True)
)
df["lead_genres"]=df["genres"].str.split(",").str[0]
df["genres_cnt"] = df["genres"].str.split(",").str.len()
df["lead_genres"]

```

```python
df.drop(['genres','production_countries'],axis=1,inplace =True)
```

```python
df.shape
df.head()
```

```python
df['seasons']=df['seasons'].fillna(0)
df['imdb_score']=df['imdb_score'].fillna(df['imdb_score'].median())
df['imdb_votes']=df['imdb_votes'].fillna(df['imdb_votes'].median())
df.dropna(inplace=True)
```

```python
dummies= pd.get_dummies(df[['type','lead_production','lead_genres']],drop_first =True)
df_dum = pd.concat([df,dummies],axis=1)
df_dum.drop(['type','lead_production','lead_genres'],axis=1,inplace=True)
df_dum.head()
```

```python
df_dum.drop('description',axis=1,inplace =True)
df_dum.head()
df_dum.set_index('title', inplace=True)
```

```python
df_dum.set_index('title', inplace=True)
```

```python
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_dum)
df_scaled =pd.DataFrame(df_scaled, columns =df_dum.columns)
df_scaled
```

```python
eps_array = [0.2, 0.5, 1]  
min_samples_array = [5, 10, 30]  

for eps in eps_array:
    for min_samples in min_samples_array:
        
        clusterer = DBSCAN(eps=eps, min_samples=min_samples).fit(df_scaled)
        cluster_labels = clusterer.labels_
        
       
        if len(set(cluster_labels)) == 1:
            continue 
        
        silhouette_avg = silhouette_score(df_scaled, cluster_labels)
        
        
        print("For eps =", eps,
              "For min_samples =", min_samples,
              "Count clusters =", len(set(cluster_labels)),
              "The average silhouette_score is :", silhouette_avg)
```

```python
dbscan_model = DBSCAN(eps=0.5, min_samples=5).fit(df_scaled)
print("For eps =", 0.5,
      "For min_samples =", 5,
      "Count clusters =", len(set(dbscan_model.labels_)),
      "The average silhouette_score is :", silhouette_score(df_scaled, dbscan_model.labels_))
```

```python
df['dbscan_clusters']= dbscan_model.labels_
df
```

```python
from thefuzz import process, fuzz

def recommend_movie(movie_name: str):
    movie_name = movie_name.lower().strip()
    all_titles = df['title'].tolist()

    # 1. Try to find the best match using Fuzzy Matching
    # extractOne returns (title, score, index)
    best_match_info = process.extractOne(movie_name, all_titles, scorer=fuzz.token_set_ratio)
    
    # We only accept a match if the score is high (e.g., > 70)
    if best_match_info and best_match_info[1] > 70:
        found_title = best_match_info[0]
        print(f"I couldn't find an exact match for '{movie_name}', but I found '{found_title}'.")
        
        # 2. Get the cluster for this found title
        movie_row = df[df['title'] == found_title].iloc[0]
        cluster = movie_row['dbscan_clusters']
        
        if cluster == -1:
            print("This movie is in our 'Unique' cluster. We don't have similar ones yet!")
            return

        # 3. Get recommendations from that cluster
        cluster_movies = df[(df['dbscan_clusters'] == cluster) & 
                                   (df['title'] != found_title)]
        
        recs = cluster_movies['title'].sample(min(len(cluster_movies), 5)).tolist()

        print(f"--- Because you liked {found_title} ---")
        for r in recs:
            print(f"Recommendation: {r}")
            
    else:
        print("Sorry, that movie is not in our database and no close matches were found.")
```

```python
s = input('Input movie name: ')

print("\n\n")
recommend_movie(s)
```
