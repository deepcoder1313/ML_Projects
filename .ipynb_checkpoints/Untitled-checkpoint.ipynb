```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_netflix =pd.read_csv("credits.csv",encoding = "latin1")
df_amazon=pd.read_csv("titles.csv",encoding = "latin1")
df_hbo =pd.read_csv("titles1.csv",encoding = "latin1")
df = pd.concat([df_netflix,df_amazon,df_hbo],axis=0)
df.head()
```

```python
df_movies = df.drop_duplicates()
df_movies
```

```python
df_movies = df.drop(["person_id","id","age_certification","imdb_id","imdb_votes","tmdb_score"],axis =1)
df_movies
```

```python
df_movies["production_countries"] = (
    df_movies["production_countries"]
    .str.replace(r"\[", "", regex=True)
    .str.replace(r"'", "", regex=True)
    .str.replace(r"\]", "", regex=True)
)
df_movies["lead_production"]=df_movies["production_countries"].str.split(",").str[0]

```

```python
df_movies["lead_production"] = df_movies["lead_production"].replace('',np.nan)
df_movies["lead_production"] 
```

```python
df_movies['production_countries'] = df_movies['production_countries'].str.split(',').str.len()
df_movies['production_countries']
```
